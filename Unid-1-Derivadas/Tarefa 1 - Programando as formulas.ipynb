```python
import numpy as np
```

```python
f = lambda x: np.sqrt(np.exp((3*x))+4*x**2) # função dada na questão
x = 2 # ponto onde vamos calcular a derivada
c = 4 # quantidade de casas decimais nos pontos f_i

# Formula obtida pela expansão da Serie de Taylor:
f2_T = lambda x, Dx: (1/(84 * ((Dx)**2))) * (-8*f(x-3*Dx) +41*f(x-2*Dx) -8*f(x-1*Dx) -50*f(x) -8*f(x+1*Dx) +41*f(x+2*Dx) -8*f(x+3*Dx))

print('--------------------------- Taylor --------')
antDf = np.Infinity
for i in range(1, 6):
  Dx, Df= 0.5**i, f2_T(x, 0.5**i)
  erro = abs(((Df-antDf)/(Df)))
  print(f'{i} --- Delta, f2T, erro --  {Dx} & {round(Df, 8)} & {round(erro, 7)}')
  print(f'{round(f(x-3*Dx), c)}, {round(f(x-2*Dx), c)}, {round(f(x-1*Dx), c)}, {round(f(x), c)}, {round(f(x+1*Dx), c)}, {round(f(x+2*Dx), c)}, {round(f(x+3*Dx), c)}')
  print('----------------')
  antDf = Df


# Formula pelos Polinomios de Interpolação de Newton:
f2_N = lambda x, Dx: (1/((Dx)**2))*((-5/48)*f(x-2.5*Dx) + (13/16)*f(x-1.5*Dx) -(17/24)*f(x-0.5*Dx)
                                    -(17/24)*f(x+0.5*Dx) +(13/16)*f(x+1.5*Dx) -(5/48)*f(x+2.5*Dx))

print('-------------------------- Newton --------')
antDf = np.Infinity
for i in range(1, 6):
  Dx, Df= 0.5**i, f2_N(x, 0.5**i)
  erro = abs(((Df-antDf)/(Df)))
  print(f'{i} --- Delta, f2N, erro --  {Dx} &    {round(Df, 8)}    & {round(erro, 7)}')
  print(f'{round(f(x-2.5*Dx), c)}, {round(f(x-1.5*Dx), c)}, {round(f(x-0.5*Dx), c)}, {round(f(x+0.5*Dx), c)}, {round(f(x+1.5*Dx), c)}, {round(f(x+2.5*Dx), c)}')
  print('----------------')
  antDf = Df
```
