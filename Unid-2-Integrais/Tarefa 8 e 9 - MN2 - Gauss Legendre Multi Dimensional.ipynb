### Tarefa 8: Multi-Dimensional - Area de superficie

 - Aplicar Gauss Legedre a um problema multdimensional.

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
```

$\int_s d S = 40^2 \frac{\pi}{2} \int^{1}_{-1} \int^{1}_{-1} \left(\sqrt{256 (0.5+\alpha/2)^2 +1}\right) (0.5+\alpha/2)d\beta d\alpha $
$ \approx  40^2 \frac{\pi}{2} \sum^{3}_{i=1} \sum^{3}_{j=1} \left[ w_i w_j \left(\sqrt{256 (0.5+\alpha_j/2)^2 +1}\right) (0.5+\alpha_j/2) \right] $

```python
# Função
f = lambda a, b: ((256*(0.5+a*0.5)**2+1)**0.5)*(0.5+a*0.5)
alphas = [-(3/5)**0.5, 0, (3/5)**0.5 ]
ws     = [5/9, 8/9, 5/9]
dt = {'(alpha_j, beta_i)': [[aj, bi] for bi in alphas 
                                        for aj in alphas],
      'w_j*w_i':[wj*wi for wi in ws
                        for wj in ws],
      'g(alpha_j, beta_i)':[],
      'w_j*w_i * g(alpha_j, beta_i)':[]
      }
S=0
for a, b in dt['(alpha_j, beta_i)']:
  dt['g(alpha_j, beta_i)'].append(f(a, b))
for g, w in zip(dt['g(alpha_j, beta_i)'],dt['w_j*w_i']):
  dt['w_j*w_i * g(alpha_j, beta_i)'].append(g*w)
S = sum(dt['w_j*w_i * g(alpha_j, beta_i)'])

dt['(alpha_j, beta_i)']=[f'({a:7.4f}, {b:7.4f})' 
                         for a, b in dt['(alpha_j, beta_i)']]
print(tabulate(dt, headers=dt.keys(), tablefmt='latex',
               colalign=['center' for _ in dt], floatfmt='.4f'))
print('Soma wjwi g() = ',S)
print('I = ',S*(40**2)*(0.5*np.pi))
```

### Tarefa 9: Volume

$400 \pi \int_{-1}^{1} \int_{-1}^{1} (0.5+0.5\alpha)^3 (320 \cos^2(\pi + \pi \beta) - 80 \sin^2(\pi +\pi\beta)) d\beta d\alpha \approx$
$\approx 400 \pi \sum^{3}_{i=1} \sum^{3}_{j=1} \left[ w_i w_j \left(320 \cos^2(\pi + \pi \beta_i) - 80 \sin^2(\pi +\pi\beta_i)\right)(0.5+0.5\alpha_j)^3 \right]$

```python
# Função
g = lambda a, b: ((1600*np.cos(np.pi/4 + b*np.pi/4)**2 -
                  400*np.sin(np.pi/4 + b*np.pi/4)**2)*(0.5+a*0.5)**3)

alphas = [-(3/5)**0.5, 0, (3/5)**0.5 ]
ws     = [5/9, 8/9, 5/9]
dt = {'(alpha_j, beta_i)': [[aj, bi] for bi in alphas 
                                        for aj in alphas],
      'w_j*w_i':[wj*wi for wi in ws
                        for wj in ws],
      'g(alpha_j, beta_i)':[],
      'w_j*w_i * g(alpha_j, beta_i)':[]
      }
S=0
for a, b in dt['(alpha_j, beta_i)']:
  dt['g(alpha_j, beta_i)'].append(g(a, b))
for gab, w in zip(dt['g(alpha_j, beta_i)'],dt['w_j*w_i']):
  dt['w_j*w_i * g(alpha_j, beta_i)'].append(gab*w)
S = sum(dt['w_j*w_i * g(alpha_j, beta_i)'])

dt['(alpha_j, beta_i)']=[f'({a:7.4f}, {b:7.4f})' 
                         for a, b in dt['(alpha_j, beta_i)']]
print(tabulate(dt, headers=dt.keys(), tablefmt='latex',
               colalign=['center' for _ in dt], floatfmt='.4f'))
print('Soma wjwi g() = ',S)
print('I = ',S*80*np.pi)
```
